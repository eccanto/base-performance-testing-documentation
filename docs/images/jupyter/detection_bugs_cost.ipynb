## Import and configure packages

```python
from pathlib import Path

import pandas
import matplotlib
from matplotlib import style


style.use('dark_background')

COLOR = '#CACACA'
matplotlib.rcParams['text.color'] = COLOR
matplotlib.rcParams['axes.labelcolor'] = COLOR
matplotlib.rcParams['xtick.color'] = COLOR
matplotlib.rcParams['ytick.color'] = COLOR

matplotlib.rcParams['font.size'] = 10
```

## Define variables

```python
# output
OUTPUT_PATH = Path('./output/')
OUTPUT_PATH.mkdir(parents=True, exist_ok=True)
```

## Plot general graph

```python
data_frame = pandas.DataFrame(
    {
        'stage': [
            'Requirements/\nArchitecture',
            'Coding',
            'Integration/\nComponent testing',
            'System/\nAcceptance testing',
            'Production/\nPost-release',
        ],
        'value': [
            2,
            6,
            10,
            15,
            30,
        ]
    }
)
plot = data_frame.plot.bar(
    x='stage',
    y='value',
    color=['limegreen', 'limegreen', 'orange', 'orange', 'orangered'],
    rot=0,
    legend=False,
    figsize=(10, 5),
)
plot.set_yticks(range(0, 30 + 1, 5))
plot.set_yticklabels([f'{index}x' for index in range(0, 30 + 1, 5)])
plot.set_xlabel('')
plot.spines[['right', 'top']].set_visible(False)
plot.spines[['left', 'bottom']].set_color(COLOR)
plot.text(-0.3, 29, 'Relative cost to fix bugs based on time of detection', fontsize=14)
plot.get_figure().savefig(OUTPUT_PATH.joinpath('detection-bugs-cost.png'), bbox_inches='tight')
```
