## Import and configure packages

```python
from pathlib import Path

import pandas
import matplotlib
from matplotlib import style


style.use('dark_background')

COLOR = '#CACACA'
matplotlib.rcParams['text.color'] = COLOR
matplotlib.rcParams['axes.labelcolor'] = COLOR
matplotlib.rcParams['xtick.color'] = COLOR
matplotlib.rcParams['ytick.color'] = COLOR

matplotlib.rcParams['font.size'] = 10
```

## Define variables

```python
# data
NORMAL_CAPACITY_DATA = [1] * 10
MAXIMUM_CAPACITY_DATA = [3] * 10
LOAD_TESTING_DATA = [0] + ([2] * 8) + [0]
STRESS_TESTING_DATA = [i * 0.38 for i in range(10)]
SOAK_TESTING_DATA = [1.5] * 10
SPIKE_TESTING_DATA = [0.97, 0.97, 3.5] + [0.97] * 7

# colors
NORMAL_CAPACITY_COLOR = 'g'
MAXIMUM_CAPACITY_COLOR = 'r'
LOAD_TESTING_COLOR = 'b'
STRESS_TESTING_COLOR = 'm'
SOAK_TESTING_COLOR = 'c'
SPIKE_TESTING_COLOR = 'y'

# output
OUTPUT_PATH = Path('./output/')
OUTPUT_PATH.mkdir(parents=True, exist_ok=True)
```

## Plot general graph

```python
def plot_graph(data, styles, output):
    data_frame = pandas.DataFrame(data)

    plot = data_frame.plot.line(
        style=styles,
        figsize=(10, 5),
    )
    plot.grid(color='#1C1C1C')
    plot.set_xticks(range(10))
    plot.set_xticklabels(['' for _ in range(10)])
    plot.set_yticks(range(4))
    plot.set_yticklabels(['' for _ in range(4)])
    plot.legend(bbox_to_anchor=(1.0, 1.0))
    plot.set_ylabel('System load')
    plot.set_xlabel('Time')
    plot.spines[['right', 'top']].set_visible(False)
    plot.spines[['left', 'bottom']].set_color(COLOR)
    plot.get_figure().savefig(output, bbox_inches='tight')


plot_graph(
    {
        'Normal capacity': NORMAL_CAPACITY_DATA,
        'Maximum capacity': MAXIMUM_CAPACITY_DATA,
        'Load testing': LOAD_TESTING_DATA,
        'Stress testing': STRESS_TESTING_DATA,
        'Soak testing': SOAK_TESTING_DATA,
        'Spike testing': SPIKE_TESTING_DATA,
    },
    [
        f'{NORMAL_CAPACITY_COLOR}-',
        f'{MAXIMUM_CAPACITY_COLOR}-',
        f'{LOAD_TESTING_COLOR}--',
        f'{STRESS_TESTING_COLOR}--',
        f'{SOAK_TESTING_COLOR}--',
        f'{SPIKE_TESTING_COLOR}--',
    ],
    OUTPUT_PATH.joinpath('performance-testing-types.png'),
)
```

## Plot load testing graph

```python
plot_graph(
    {
        'Normal capacity': NORMAL_CAPACITY_DATA,
        'Maximum capacity': MAXIMUM_CAPACITY_DATA,
        'Load testing': LOAD_TESTING_DATA,
    },
    [
        f'{NORMAL_CAPACITY_COLOR}-',
        f'{MAXIMUM_CAPACITY_COLOR}-',
        f'{LOAD_TESTING_COLOR}--',
    ],
    OUTPUT_PATH.joinpath('load-testing.png'),
)
```

## Plot stress testing graph

```python
plot_graph(
    {
        'Normal capacity': NORMAL_CAPACITY_DATA,
        'Maximum capacity': MAXIMUM_CAPACITY_DATA,
        'Stress testing': STRESS_TESTING_DATA,
    },
    [
        f'{NORMAL_CAPACITY_COLOR}-',
        f'{MAXIMUM_CAPACITY_COLOR}-',
        f'{STRESS_TESTING_COLOR}--',
    ],
    OUTPUT_PATH.joinpath('stress-testing.png'),
)
```

```python
plot_graph(
    {
        'Normal capacity': NORMAL_CAPACITY_DATA,
        'Maximum capacity': MAXIMUM_CAPACITY_DATA,
        'Soak testing': SOAK_TESTING_DATA,
    },
    [
        f'{NORMAL_CAPACITY_COLOR}-',
        f'{MAXIMUM_CAPACITY_COLOR}-',
        f'{SOAK_TESTING_COLOR}--',
    ],
    OUTPUT_PATH.joinpath('soak-testing.png'),
)
```
